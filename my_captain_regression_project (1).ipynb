```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston
```

```python
#understanding the dataset
boston = load_boston()
```

```python
boston
```

```python
print(boston.DESCR)
```

```python
# acess data attributes
dataset=boston.data
for name,index in enumerate(boston.feature_names):
    print(index,name)
```

```python
# reshaping data
data=dataset[:,12].reshape(-1,1)
data
```

```python
data.shape
```

```python
np.shape(dataset)
```

```python
#target value
target = boston.target.reshape(-1,1)
target
```

```python
np.shape(target)
```

```python
%matplotlib inline
```

```python
#ensuring matplotlib is working
plt.scatter(data, target,color="green")
plt.xlabel("LOWER INCOME POPULATION")
plt.ylabel("COST OF HOUSE")
plt.show()
```

# Linear Regression model
- linear regression
- lasso
- ridge

```python
#linear regression 
from sklearn.linear_model import LinearRegression

#creating a regression model
reg = LinearRegression()

#fit the mode
reg.fit(data,target)
```

```python
#predicting 
pred = reg.predict(data)
```

```python
#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```

```python
# laso
from sklearn.linear_model import Lasso

#creating a regression model
reg = Lasso()

#fit the mode
reg.fit(data,target)

#predicting 
pred = reg.predict(data)

#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```

```python
# ridge model
from sklearn.linear_model import Ridge

#creating a regression model
reg = Ridge()

#fit the mode
reg.fit(data,target)

#predicting 
pred = reg.predict(data)

#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```

# ploynomial model

```python
#circuventing curve issue using polynomial model
from sklearn.preprocessing import PolynomialFeatures

#to allow merging  of models 
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(PolynomialFeatures(3),reg)
```

```python
model.fit(data,target)
```

```python
pred =model.predict(data)
```

```python
#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```

```python
# R_2 metric
from sklearn.metrics import r2_score
```

```python
# predict
r2_score(pred,target)
```

# SVM 

```python
# SVM.Svr
from sklearn import svm

#creating a regression model
reg = svm.SVR()

#fit the mode
reg.fit(data,target)

#predicting 
pred = reg.predict(data)

#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```

```python
# svm.LinearSVR
from sklearn import svm

#creating a regression model
reg = svm.LinearSVR()

#fit the mode
reg.fit(data,target)

#predicting 
pred = reg.predict(data)

#ensuring the matplotlib is working 
plt.scatter(data,target,color='green')
plt.plot(data,pred,color='red')
plt.xlabel("Lower income population")
plt.ylabel("Cost of house")
plt.show()
```
